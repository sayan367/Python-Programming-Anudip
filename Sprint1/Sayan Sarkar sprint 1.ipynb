```python

custom_colors = ['yellow', 'blue', 'green', 'red']  # Add more colors as needed

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='category', palette=custom_colors)  # Use your custom color list
plt.title('Distribution of Product Categories')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python

# 2. Brand-wise Sales (using sum of amount)

# Set the figure size for the plot
plt.figure(figsize=(10,6))

# Group the DataFrame by 'brand', summing the 'amount' for each brand,
# sort the results in descending order, and select the top 10 brands
brand_sales = df.groupby('brand')['amount'].sum().sort_values(ascending=False).head(10)

# Create a bar plot using seaborn to visualize the total sales amount for the top 10 brands
sns.barplot(x=brand_sales.index, y=brand_sales.values, palette='muted')

# Set the title of the plot
plt.title('Top 10 Brands by Total Sales Amount')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Adjust the layout to make sure everything fits well in the figure
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# 5. Pie Chart: Gender distribution
plt.figure(figsize=(6, 6))
gender_count = df['gender'].value_counts()
plt.pie(gender_count, labels=gender_count.index, autopct='%1.1f%%', startangle=90, colors=['green','yellow'])
plt.title('Gender Distribution')
plt.tight_layout()
plt.show()
```

```python
# 6. Line Chart: Sales trend over years
plt.figure(figsize=(10, 6))
yearly_sales = df.groupby('year')['amount'].sum()
sns.lineplot(x=yearly_sales.index, y=yearly_sales.values, marker='o')
plt.title('Sales Trend Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Sales (Amount)')
plt.tight_layout()
plt.show()
```

```python
# 7. Histogram: Distribution of Ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['rating'], bins=5, kde=True, color='green')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#Stacked Bar Chart: Comparison of categories across genders

plt.figure(figsize=(12, 7)) 
category_gender = df.groupby(['category', 'gender']).size().unstack()

# Create the stacked bar chart
category_gender.plot(kind='bar', stacked=True, color=['blue', 'red'], figsize=(12, 7))

# Title and labels
plt.title('Comparison of Item Categories by Gender')
plt.xlabel('Category')
plt.ylabel('Count')

# Adjust x-axis label rotation and alignment
plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels by 30 degrees and align to the right

# Adjust layout to make sure everything fits
plt.tight_layout()

# Show the plot
plt.show()
```

```python
**Total Sales per Month:** Displays the total sales in each month to identify peak periods.  
**Average Purchase Amount per Month:** Shows the average amount spent per purchase in each month.  
**Total Quantity Sold per Month:** Highlights the total number of units sold in each month.
```

```python
# Assuming 'data' is already loaded as a DataFrame

# Create subplots for Monthly Sales Analysis
fig, axes = plt.subplots(3, 1, figsize=(12, 12))

# 1. Total Sales per Month
monthly_sales = df.groupby('month')['amount'].sum()
axes[0].plot(monthly_sales.index, monthly_sales.values, marker='o', color='Red')
axes[0].set_title('Total Sales per Month')
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Total Sales (Amount)')
axes[0].set_xticks(range(1, 13))
axes[0].set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# 2. Average Purchase Amount per Month
avg_purchase_per_month = df.groupby('month')['amount'].mean()
axes[1].plot(avg_purchase_per_month.index, avg_purchase_per_month.values, marker='s', color='green')
axes[1].set_title('Average Purchase Amount per Month')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Average Purchase Amount')
axes[1].set_xticks(range(1, 13))
axes[1].set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# 3. Total Quantity Sold per Month
total_quantity_per_month = df.groupby('month')['quantity'].sum()
axes[2].bar(total_quantity_per_month.index, total_quantity_per_month.values, color='Green')
axes[2].set_title('Total Quantity Sold per Month')
axes[2].set_xlabel('Month')
axes[2].set_ylabel('Total Quantity Sold')
axes[2].set_xticks(range(1, 13))
axes[2].set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Adjust layout for better spacing
plt.tight_layout()

# Show the subplot
plt.show()
```

```python
#**Total Sales per Month:** Displays the total sales in each month to identify peak periods.  
#**Average Purchase Amount per Month:** Shows the average amount spent per purchase in each month.  
#**Total Quantity Sold per Month:** Highlights the total number of units sold in each month.
```
